# <b style="color: Red">Loops: </b>
<b><p>Loops let you repeat the code writing over again and again (multiple times until you set some limit)</p></b>
<b><p>Loops can be divided into 2 types: 
- 'for' loop and 'while' loop.</p></b>

## <b style="color: Red">For Loop: </b>
<b><p>For loops can be used when you know how many times you have to repeat the code.</p>

## <b style="color: Red">While Loop: </b>
<b><p>While loops can be used when you want to repeat until something changes.</p>

```python
# For loop
fruits = ["apple", "orange", "pineapple", "mango"]

for fruit in fruits:
    print(fruit)
```

```python
# Example:
for i in range(5):
    print(i)
```

```python
# Example 2:
for i in range(0,10,2):
    print(i) # It will print the even number from 0 to 10
```

```python
# Example 3:
word = "SUMAN"
for letters in word:
    print(letters)
```

# <b style = "color : red"> Level 2 of loops:
<p><b>Practising the level 2 of the loops concept</p>

```python
# WAP to count how many digits are there in a number
num = int(input("Enter your given number: "))
count = 0

while num > 0:
    num = num // 10
    count = count + 1

print(f"Number of digits: {count}")
```

```python
# WAP to find the sum of digits given by the user
num = int(input("Enter a number: "))
total = 0

while num > 0:
    digit = num % 10 
    total = total + digit
    num = num // 10
    
print(f"Sum of digits: {total}")
```

```python
# WAP to find the product of digits of a number

num = int(input("Enter your number: "))
product = 0

while num > 0:
    digit = num % 10
    product = product * digit
    num = num // 10

print(f"Prodcut of digits: {product}")
```

# <b style = "color: red"> Level 3 Loops

```python
# Q21. Right-Angle Triangle of Stars
rows = 5

for i in range (1, rows + 1):
    for j in range(i):
        print("*", end = "")
    print()
```

```python
# Q22. Inverted Right-Angle Triangle
rows = 5

for i in range (rows, 0, -1):
    for j in range(i):
        print("*", end = "")
    print()
```

```python
# Q23. Pyramid of Stars
rows = 5

for i in range(1, rows + 1):
    # Print space
    for space in range(rows - i):
        print(" ", end = "")

    # Print stars
    for star in range(2 * i - 1):
        print("*", end = "")

    print()
```
