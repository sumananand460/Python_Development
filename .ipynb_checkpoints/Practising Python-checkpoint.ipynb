````python
# Using lambda function filter even numbers using filter()
# 1. Accept space-separated numbers from the user
user_input = input("Enter numbers separated by spaces: ")

# 2. Convert the input string into a list of integers
numbers = list(map(int, user_input.split()))

# 3. Filter out the even numbers
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

# 4. Print the final results
print(f"Your original list: {numbers}")
print(f"The even numbers are: {even_numbers}")
````

# <b style = "color: red">Recursion

### Calling the function again and again 

````python
# WAP to print numbers from 1 to n using recursion
def print_numbers(n):
    if n == 0:
        return
    
    print_numbers(n-1)
    print(n, end = " ")

print_numbers(10)
````

````python
# WAP to print numbers from n to 1
def reverse_numbers(n):
    if n == 0:
        return

    print(n, end= " ")
    reverse_numbers(n-1)

reverse_numbers(10)
````

````python
# WAP to print the factorial using recursion
n = int(input("Enter your number:"))
def factorial(n):
    
    # BASE CASE
    if n == 1:
        return 1

    # RECURSIVE CASE
    return n * factorial(n-1)

print(f"Factorial of {n} is {factorial(n)}")
````

````python
# WAP using recursion to print the sum of first n natural numbers
n = int(input("Enter your number:"))
def total(n):

    # Base Case
    if n == 1:
        return 1

    # Recursive Case
    return n + total(n-1)

print(f"Sum: {total(n)}")
````

# <b style = "color: red">Object Oriented Programming

````python
# 1. THE CLASS (The Blueprint)
class Dog:
    # This sets up the characteristics (Attributes) when a dog is born
    def __init__(self, name, breed):
        self.name = name    # Characteristic 1
        self.breed = breed  # Characteristic 2

    # This is an action (Method) the dog can perform
    def bark(self):
        print(f"{self.name} says Woof!")

# ---------------------------------------------------------

# 2. THE OBJECTS (The Real Things)
# We use the blueprint to create two real, distinct dogs
dog1 = Dog("Buddy", "Golden Retriever")
dog2 = Dog("Max", "German Shepherd")

# 3. USING THE OBJECTS
print(dog1.name)   # Output: Buddy
dog2.bark()        # Output: Max says Woof!
````

````python
# Using OOPs to show class and object
class Student:
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks

# Creating a method to call later
    def display(self):
        print(f"{self.name} : {self.marks}")

# Creating Objects
s1 = Student("Arya", 87)
s2 = Student("Anand", 92)

# Calling the method
display(s1)
display(s2)
````

````python
'''
Question 1: The Bank Account (Beginner)
Task: Create a class named BankAccount.
It should have two attributes: owner (a string) and balance (an integer/float).
Inside the class, create a method called deposit(amount) that adds money to the balance.
Create an object of this class, deposit $50, and print the updated balance.
'''
# Solution
class BankAccount:
    def __init__(self,owner,balance):
        self.owner = owner
        self.balance = balance 

    def deposit(self,amount):
        self.balance += amount

account = BankAccount("Bruce",100000)
account.deposit(500)
print(f"Updated balance: ${account.balance}")
````

# <b style = "color: red">Inheritance
### Inheritance in Python is a core Object-Oriented Programming (OOP) concept that allows one class (called a child or subclass) to acquire the properties and methods of another class (called a parent or superclass). It promotes code reuse and establishes a natural hierarchy between classes.

````python
# Parent Class
class Animal:
    def __init__(self, name):
        self.name = name
    def speak(self):
        return "Some generic sound"

# Child Class
class Dog(Animal): # Child class inherits from parent class (Animal)
    def speak(self): # Overrride parent method
        return f"{self.name} says Woof!"

class Cat(Animal): # Another child class
    def speak(self):
        return f"{self.name} says Meow!"

# Usage
dog = Dog("Buddy")
cat = Cat("Whiskers")

print(dog.speak()) # Buddy says Woof
print(cat.speak()) # Whiskers says Meow
````

# Encapsulation

````python
class Student:
    def __init__(self, name, marks):
        self.__name = name       # private variable
        self.__marks = marks     # private variable

    # Instead of getter/setter, we can use a method directly
    def display(self):
        print(f"{self.__name} : {self.__marks}")

# Usage
s1 = Student("Arya", 87)
s1.display()

# Direct access will fail
print(s1.__marks)   # ❌ Error: Attribute not accessible
````

# Method Overriding
<p><b>Method Overriding refers to a child class redefining a parent class method.
<li>Method Overriding also means run-time polymorphism </p>

````python
# Example of method overrriding
class Animal:
    def speak(self):
        print("Some generic sound!")

class Cat:
    def speak(self):
        print("Meow!")

class Dog:
    def speak(self):
        print("Bark!")

class Cow:
    def speak(self):
        print("Mooh!")

animals = [Cat(), Dog(), Cow()]

for animal in animals:
    animal.speak()
````

# Method Overloading
<p><b>Method overloading means having multiple functions with the exact same name, but they accept a different number or type of inputs.
<li>Also knows as Compile -time polymorphism </li></p>

````python
# Example of method overloading
class Calculator:
    def add(self, a, b=0, c=0):
        return a + b + c

calc = Calculator()

print(calc.add(5))
print(calc.add(5,4,20))
print(calc.add(0,34,6))
print(calc.add(0))
````

````python
# Example of method overloading with the help of for-loop also using *args
class Calculator:
    def add(self, *numbers):
        total = 0
        for num in numbers:
            total += num
        return total

calc = Calculator()

print(calc.add(5,6))
````

# Python Logging - Complete Notes for Beginners

## What is Logging?

Logging is the process of recording events that happen while a program runs.

Instead of using:

```python
print("Program Started")
```

Professional developers use:

```python
logging.info("Program Started")
```

### Why Logging?

* Debug errors
* Track program execution
* Monitor applications
* Store events in files
* Help during maintenance

---

## Importing Logging Module

### Code Cell

```python
import logging
```

---

## Basic Logging Configuration

The `basicConfig()` function is used to configure logging.

### Code Cell

```python
import logging

logging.basicConfig(level=logging.INFO)

logging.info("Application Started")
```

### Output

```text
INFO:root:Application Started
```

---

## Logging Levels

Python provides five main logging levels.

| Level    | Purpose              |
| -------- | -------------------- |
| DEBUG    | Detailed information |
| INFO     | General information  |
| WARNING  | Something unexpected |
| ERROR    | Serious problem      |
| CRITICAL | Program may stop     |

---

## Example of All Logging Levels

### Code Cell

```python
import logging

logging.basicConfig(level=logging.DEBUG)

logging.debug("Debug Message")
logging.info("Info Message")
logging.warning("Warning Message")
logging.error("Error Message")
logging.critical("Critical Message")
```

### Output

```text
DEBUG:root:Debug Message
INFO:root:Info Message
WARNING:root:Warning Message
ERROR:root:Error Message
CRITICAL:root:Critical Message
```
